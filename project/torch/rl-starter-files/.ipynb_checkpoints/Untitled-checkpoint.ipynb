```python
def ret_res(t,folder):
    trains = []
    evals = []
    for i in range(1,6):
        dat = pd.read_csv("/Users/anandsranjan/Desktop/AsymmPlay/project/torch/rl-starter-files/"+folder+"/"+t+"_"+str(i)+"/log.csv")
        train_avg = dat.to_numpy()[:,4]
        eval_avg = dat.to_numpy()[:,-4]
        trains.append(train_avg)
        evals.append(eval_avg)
    return trains,evals
```

```python
def run_ret(folder):
    t,e=ret_res("teaching",folder)
    t_t_avg,t_e_avg = np.average(t,axis=0),np.average(e,axis=0)
    
    t,e=ret_res("no_teaching",folder)
    nt_t_avg,nt_e_avg = np.average(t,axis=0),np.average(e,axis=0)
    
    t,e=ret_res("teaching_ha",folder)
    th_t_avg,th_e_avg = np.average(t,axis=0),np.average(e,axis=0)
    
    plt.figure()
    plt.title("train")
    plt.plot(t_t_avg, label="t")
    plt.plot(th_t_avg,label="th")
    plt.plot(nt_t_avg,label="nt")
    plt.legend()
    
    plt.figure()
    plt.title("eval")
    plt.plot(t_e_avg,label="t")
    plt.plot(th_e_avg,label="th")
    plt.plot(nt_e_avg,label="nt")
    plt.legend()
    
def plt_runs(folder,b):
    t,e=ret_res(b+"_diff",folder)
    t_t_avg,t_e_avg = np.average(t,axis=0),np.average(e,axis=0)
    
    t,e=ret_res(b+"_rand",folder)
    nt_t_avg,nt_e_avg = np.average(t,axis=0),np.average(e,axis=0)
    
    t,e=ret_res(b+"_same",folder)
    th_t_avg,th_e_avg = np.average(t,axis=0),np.average(e,axis=0)
    
    plt.figure()
    plt.title("train")
    plt.plot(t_t_avg, label="diff")
    plt.plot(th_t_avg,label="rand")
    plt.plot(nt_t_avg,label="same")
    plt.legend()
    
    plt.figure()
    plt.title("eval")
    plt.plot(t_e_avg,label="diff")
    plt.plot(th_e_avg,label="rand")
    plt.plot(nt_e_avg,label="same")
    plt.legend()
```
